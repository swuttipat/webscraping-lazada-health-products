```python
import pandas as pd

df = pd.read_csv('health_and_wellness_20230802_160225.csv')

df.info()
```

```python
df.head()
```

```python
# Transform column name to lowercase and replace ' ' with '_'
df.columns = df.columns.str.replace(' ', '_').str.lower()

# Fill missing values
df['total_sold'].fillna('0', inplace=True)
df['total_reviews'].fillna('0', inplace=True)
df['shop_location'].fillna('Unknown', inplace=True)

# Remove non-numeric characters using regex and convert to appropriate data types
df['price'] = df['price'].str.replace('‡∏ø', '', regex=True).str.replace(',', '', regex=True).astype(float)

# Cleaning 'total_sold' column
df['total_sold'] = df['total_sold'].str.replace('[^0-9.]+', '', regex=True)
df['total_sold'] = df['total_sold'].replace('9999+', '10000').astype(int)

# Cleaning 'total_reviews' column
df['total_reviews'] = df['total_reviews'].str.replace('[^0-9.]+', '', regex=True).astype(int)

# Display the first few rows of the cleaned dataframe
df.head()
```
